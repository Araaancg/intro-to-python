## 1. ¿Qué es la recursión?

**Recursividad**: Definir una operación **en términos de sí misma**.

En programación: una función que **se llama a sí misma** para resolver problemas dividiéndolos en partes más pequeñas.

### 1.1 La analogía del aula

Imagina que quieres contar cuántos estudiantes hay en una columna, pero **solo puedes ver al de inmediatamente detrás**.

**Solución recursiva:**
1. Pregunta al de detrás: "¿Cuántos hay a partir de ti?"
2. Ese estudiante hace lo mismo con el siguiente
3. El último (no tiene a nadie detrás) responde: "1" (solo yo)
4. Cada uno suma 1 a la respuesta que recibe y la devuelve

```
Estudiante A → pregunta → Estudiante B → pregunta → Estudiante C (último)
                                                           ↓
                                                     devuelve 1
                                    ↓
                              devuelve 1+1 = 2
         ↓
   devuelve 2+1 = 3
```

### 1.2 Los dos ingredientes de toda recursión

| Componente | Descripción | Si falta... |
|------------|-------------|-------------|
| **Caso base** | Problema trivial que se resuelve directamente | Recursión infinita → 💥 |
| **Caso recursivo** | Divide el problema y se llama a sí misma | La función no hace nada útil |

## 2. Primer ejemplo: Invertir un string

Queremos invertir un texto: `"hola"` → `"aloh"`

### Pensamiento recursivo:

Para invertir `"hola"`:
1. Tomo el primer carácter: `"h"`
2. Invierto el resto: `"ola"` → `"alo"`
3. Pongo el primer carácter **al final**: `"alo"` + `"h"` = `"aloh"`

**Caso base:** Un string vacío o de 1 carácter ya está "invertido".

````python
def invertir_string(texto: str) -> str:
    """Invierte un string de forma recursiva."""
    
    # CASO BASE: string vacío o de 1 carácter
    if len(texto) == 1:
        return texto
    
    # CASO RECURSIVO: invertir el resto + primer carácter al final
    primer_caracter = texto[0]
    resto = texto[1:]
    
    return invertir_string(resto) + primer_caracter


# Probamos
print(f"'hola' → '{invertir_string('hola')}'")
print(f"'Python' → '{invertir_string('Python')}'")
print(f"'a' → '{invertir_string('a')}'")
print(f"'' → '{invertir_string('')}'")
````

### Traza de ejecución paso a paso

Veamos qué pasa cuando llamamos a `invertir_string("hola")`:

```
invertir_string("hola")
  │ primer_caracter = "h", resto = "ola"
  └─> invertir_string("ola") + "h"
        │ primer_caracter = "o", resto = "la"
        └─> invertir_string("la") + "o"
              │ primer_caracter = "l", resto = "a"
              └─> invertir_string("a") + "l"
                    │
                    └─> return "a"  ← CASO BASE (1 carácter)
              │
              └─> return "a" + "l" = "al"
        │
        └─> return "al" + "o" = "alo"
  │
  └─> return "alo" + "h" = "aloh"
```

````python
# Versión con prints para ver cada paso

def invertir_string_debug(texto: str, nivel: int = 0) -> str:
    """Versión con debug para ver la recursión."""
    sangria = "  " * nivel
    print(f"{sangria}invertir_string('{texto}')")
    
    # Caso base
    if len(texto) <= 1:
        print(f"{sangria}  → Caso base, devuelvo '{texto}'")
        return texto
    
    # Caso recursivo
    primer_caracter = texto[0]
    resto = texto[1:]
    
    resultado_resto = invertir_string_debug(resto, nivel + 1)
    resultado = resultado_resto + primer_caracter
    
    print(f"{sangria}  → '{resultado_resto}' + '{primer_caracter}' = '{resultado}'")
    return resultado


print("=" * 50)
resultado = invertir_string_debug("hola")
print("=" * 50)
print(f"Resultado final: '{resultado}'")
````

### ⚠️ Error común: olvidar reducir el problema

````python
def invertir_mal(texto: str) -> str:
    """¡ESTO ESTÁ MAL! Recursión infinita."""
    if len(texto) <= 1:
        return texto
    
    # ❌ Error: llamamos con el mismo texto, nunca llegamos al caso base
    return invertir_mal(texto) + texto[0]

# NO ejecutar esto - cuelga el programa
# invertir_mal("hola")
````

## 3. Segundo ejemplo: Dibujar un árbol de asteriscos

Queremos dibujar un patrón como este:

```
*
**
***
****
*****
```

### Pensamiento recursivo:

Para dibujar un árbol de altura 5:
1. Primero dibujo un árbol de altura 4
2. Luego dibujo la línea de 5 asteriscos

**Caso base:** Árbol de altura 0 = no dibujar nada

````python
# in: int 5
# out: arbol
````

````python
def dibujar_arbol(altura: int) -> None:
    """Dibuja un árbol de asteriscos de forma recursiva."""
    
    # CASO BASE: altura 0, no hay nada que dibujar
    if altura == 0:
        return
    
    # CASO RECURSIVO:
    # 1. Primero dibujar el árbol más pequeño (altura - 1)
    dibujar_arbol(altura - 1)
    
    # 2. Luego dibujar la línea actual
    print("*" * altura)


# Probamos
print("Árbol de altura 5:")
print("-" * 20)
dibujar_arbol(5)



# 5
# 4
# 3
# 2 -> dibujar_arbol(0) | **
# 1 -> dibujar_arbol(0) | *


````

````python
# ¿Y si queremos el árbol al revés?
# Solo cambiamos el orden: primero la línea, luego el árbol más pequeño

def dibujar_arbol_invertido(altura: int) -> None:
    """Dibuja un árbol invertido de asteriscos."""
    
    # Caso base
    if altura == 0:
        return
    
    # Caso recursivo (orden invertido)
    # 1. Primero dibujar la línea actual
    print("*" * altura)
    
    # 2. Luego dibujar el árbol más pequeño
    dibujar_arbol_invertido(altura - 1)


print("Árbol invertido de altura 5:")
print("-" * 20)
dibujar_arbol_invertido(5)
````

````python
# ¿Y un árbol centrado (tipo pino)?

def dibujar_pino(altura: int, altura_total: int = None) -> None:
    """Dibuja un pino centrado."""
    
    # Guardamos la altura total para calcular los espacios
    if altura_total is None:
        altura_total = altura
    
    # Caso base
    if altura == 0:
        return
    
    # Caso recursivo
    dibujar_pino(altura - 1, altura_total)
    
    # Calcular espacios y asteriscos
    espacios = altura_total - altura
    asteriscos = 2 * altura - 1
    print(" " * espacios + "*" * asteriscos)


print("🎄 Pino de altura 6:")
print("-" * 20)
dibujar_pino(6)
````

### ¿Por qué funciona?

El orden de las operaciones importa:

| Versión | Orden | Resultado |
|---------|-------|--------|
| Normal | Primero recursión, luego print | Crece hacia abajo |
| Invertido | Primero print, luego recursión | Decrece hacia abajo |

Es como una pila de platos: si pones primero y luego llamas a recursión, los prints "esperan" hasta que la recursión termine.

## 4. Tercer ejemplo: Explorar carpetas (aplicación real)

Este es un caso de uso **real** de recursión: explorar un sistema de archivos.

Las carpetas pueden contener:
- Archivos
- Otras carpetas (que a su vez contienen archivos y carpetas...)

¡Es recursivo por naturaleza!

### Estructura de ejemplo:
```
📁 proyecto/
    📄 readme.txt
    📁 src/
        📄 main.py
        📄 utils.py
        📁 tests/
            📄 test_main.py
    📁 docs/
        📄 manual.pdf
```

````python
# Primero, simulamos una estructura de carpetas con diccionarios

sistema_archivos = {
    "nombre": "proyecto",
    "tipo": "carpeta",
    "contenido": [
        {"nombre": "readme.txt", "tipo": "archivo"},
        {
            "nombre": "src",
            "tipo": "carpeta",
            "contenido": [
                {"nombre": "main.py", "tipo": "archivo"},
                {"nombre": "utils.py", "tipo": "archivo"},
                {
                    "nombre": "tests",
                    "tipo": "carpeta",
                    "contenido": [
                        {"nombre": "test_main.py", "tipo": "archivo"}
                    ]
                }
            ]
        },
        {
            "nombre": "docs",
            "tipo": "carpeta",
            "contenido": [
                {"nombre": "manual.pdf", "tipo": "archivo"}
            ]
        }
    ]
}
````

````python
def mostrar_estructura(elemento: dict, nivel: int = 0) -> None:
    """Muestra la estructura de carpetas de forma recursiva."""
    
    sangria = "    " * nivel
    
    if elemento["tipo"] == "archivo":
        # CASO BASE: es un archivo, solo lo mostramos
        print(f"{sangria}📄 {elemento['nombre']}")
    
    else:
        # CASO RECURSIVO: es una carpeta
        print(f"{sangria}📁 {elemento['nombre']}/")
        
        # Exploramos cada elemento dentro de la carpeta
        for item in elemento["contenido"]:
            mostrar_estructura(item, nivel + 1)  # ¡Llamada recursiva!


# Mostramos la estructura
print("Estructura del proyecto:")
print("=" * 40)
mostrar_estructura(sistema_archivos)
````

````python
def contar_archivos(elemento: dict) -> int:
    """Cuenta el número total de archivos de forma recursiva."""
    
    if elemento["tipo"] == "archivo":
        # CASO BASE: es un archivo, cuenta como 1
        return 1
    
    else:
        # CASO RECURSIVO: es una carpeta, sumar archivos de su contenido
        total = 0
        for item in elemento["contenido"]:
            total += contar_archivos(item)  # ¡Llamada recursiva!
        return total


total = contar_archivos(sistema_archivos)
print(f"Total de archivos en el proyecto: {total}")
````

````python
def buscar_archivo(elemento: dict, nombre_buscado: str, ruta_actual: str = "") -> str:
    """Busca un archivo por nombre y devuelve su ruta."""
    
    ruta = ruta_actual + "/" + elemento["nombre"] if ruta_actual else elemento["nombre"]
    
    if elemento["tipo"] == "archivo":
        # CASO BASE: es un archivo
        if elemento["nombre"] == nombre_buscado:
            return ruta
        else:
            return None
    
    else:
        # CASO RECURSIVO: es una carpeta, buscar en su contenido
        for item in elemento["contenido"]:
            resultado = buscar_archivo(item, nombre_buscado, ruta)
            if resultado:  # Si lo encontró, devolver la ruta
                return resultado
        return None  # No se encontró en esta carpeta


# Probamos
archivos_a_buscar = ["test_main.py", "manual.pdf", "noexiste.txt"]

for archivo in archivos_a_buscar:
    ruta = buscar_archivo(sistema_archivos, archivo)
    if ruta:
        print(f"✓ '{archivo}' encontrado en: {ruta}")
    else:
        print(f"✗ '{archivo}' no encontrado")
````

### ¿Por qué la recursión es perfecta aquí?

Las carpetas son **estructuras recursivas por naturaleza**:
- Una carpeta contiene archivos y **otras carpetas**
- Esas carpetas contienen archivos y **otras carpetas**
- Y así sucesivamente...

No sabemos de antemano cuántos niveles de profundidad hay. La recursión maneja esto automáticamente.

## 5. Ejemplo clásico: Factorial

El factorial de n es: `n! = n × (n-1) × (n-2) × ... × 2 × 1`

**Definición recursiva:**
- `0! = 1` (caso base)
- `n! = n × (n-1)!` (caso recursivo)

````python
def factorial(n: int) -> int:
    """Calcula el factorial de n."""
    
    # Caso base
    if n == 0 or n == 1:
        return 1
    
    # Caso recursivo
    return n * factorial(n - 1)


print(f"5! = {factorial(5)}")   # 120
print(f"0! = {factorial(0)}")   # 1
print(f"10! = {factorial(10)}") # 3628800
````

## 6. Fibonacci

La sucesión de Fibonacci: 0, 1, 1, 2, 3, 5, 8, 13, 21...

Cada número es la **suma de los dos anteriores**.

**Definición recursiva:**
- `fib(0) = 0`
- `fib(1) = 1`
- `fib(n) = fib(n-1) + fib(n-2)`

````python
def fibonacci(n: int) -> int:
    """Devuelve el n-ésimo número de Fibonacci."""
    
    # Casos base
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # Caso recursivo: DOS llamadas recursivas
    return fibonacci(n - 1) + fibonacci(n - 2)


# Primeros 10 números de Fibonacci
for i in range(10):
    print(f"fib({i}) = {fibonacci(i)}")
````

### ⚠️ Problema: Fibonacci recursivo es MUY lento

Para calcular `fib(5)`, se calcula `fib(3)` dos veces, `fib(2)` tres veces...

Se repiten muchos cálculos. En temas posteriores veremos cómo optimizar esto con **memoización** y **programación dinámica**.

````python
# Comparación de tiempos
import time

for n in [10, 20, 30, 35]:
    inicio = time.time()
    resultado = fibonacci(n)
    tiempo = time.time() - inicio
    print(f"fib({n}) = {resultado:,} → {tiempo:.3f} segundos")
````

## 7. ¿Por qué usar recursión?

### Ventajas
- Código **más elegante y legible** para ciertos problemas
- Natural para estructuras **jerárquicas** (árboles, carpetas, menús)
- Esencial para algoritmos como **divide y vencerás**

### Desventajas
- Puede ser **más lenta** que la versión iterativa
- Consume **más memoria** (cada llamada usa espacio en la pila)
- Riesgo de **stack overflow** con recursión muy profunda

````python
# Límite de recursión en Python
import sys
print(f"Límite de recursión actual: {sys.getrecursionlimit()}")

# Se puede cambiar (con cuidado):
# sys.setrecursionlimit(10000)
````

## 📝 Resumen

| Concepto | Descripción |
|----------|-------------|
| **Recursión** | Función que se llama a sí misma |
| **Caso base** | Condición de parada (¡obligatorio!) |
| **Caso recursivo** | Reduce el problema y se llama de nuevo |
| **Ejemplos ideales** | Strings, árboles, carpetas, estructuras jerárquicas |
| **Peligros** | Recursión infinita, stack overflow, ineficiencia |
