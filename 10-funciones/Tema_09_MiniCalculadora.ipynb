En este ejercicio vamos a ver c√≥mo **refactorizar** (mejorar) un programa que funciona pero est√° mal estructurado, convirti√©ndolo en un programa limpio y modular usando funciones.

---

## üçù VERSI√ìN 1: C√≥digo "espagueti" (ANTES)

Este c√≥digo **funciona**, pero tiene varios problemas:
- C√≥digo repetido
- Dif√≠cil de leer
- Dif√≠cil de mantener
- Todo mezclado

```python
# ============================================
# VERSI√ìN 1: TODO JUNTO (c√≥digo espagueti)
# ============================================

opcion = 0

while opcion != 5:
    
    # Mostrar men√∫
    print("")
    print("===== CALCULADORA =====")
    print("1. Sumar")
    print("2. Restar")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Salir")
    print("=======================")
    
    opcion = int(input("Elige opci√≥n: "))
    
    if opcion == 1:
        num1 = float(input("N√∫mero 1: "))
        num2 = float(input("N√∫mero 2: "))
        resultado = num1 + num2
        print("------------------------")
        print(f"{num1} + {num2} = {resultado}")
        print("------------------------")
        
    elif opcion == 2:
        num1 = float(input("N√∫mero 1: "))
        num2 = float(input("N√∫mero 2: "))
        resultado = num1 - num2
        print("------------------------")
        print(f"{num1} - {num2} = {resultado}")
        print("------------------------")
        
    elif opcion == 3:
        num1 = float(input("N√∫mero 1: "))
        num2 = float(input("N√∫mero 2: "))
        resultado = num1 * num2
        print("------------------------")
        print(f"{num1} * {num2} = {resultado}")
        print("------------------------")
        
    elif opcion == 4:
        num1 = float(input("N√∫mero 1: "))
        num2 = float(input("N√∫mero 2: "))
        if num2 != 0:
            resultado = num1 / num2
            print("------------------------")
            print(f"{num1} / {num2} = {resultado}")
            print("------------------------")
        else:
            print("Error: No se puede dividir entre 0")
            
    elif opcion == 5:
        print("¬°Hasta luego!")
        
    else:
        print("Opci√≥n no v√°lida")
```

### ‚ùå Problemas de esta versi√≥n:

1. **C√≥digo repetido**: `input("N√∫mero 1: ")` aparece 4 veces
2. **C√≥digo repetido**: Las l√≠neas de separaci√≥n `print("---")` aparecen 8 veces
3. **Todo mezclado**: Men√∫, l√≥gica, entrada de datos y salida est√°n juntos
4. **Dif√≠cil de modificar**: Si quiero cambiar el formato de salida, tengo que cambiarlo en 4 sitios

---

## ‚ú® VERSI√ìN 2: C√≥digo modularizado (DESPU√âS)

Vamos a aplicar dos principios:
1. **Modularizaci√≥n**: Dividir en funciones por funcionalidad
2. **No repetir c√≥digo**: Una funci√≥n para cada tarea que se repite

### Paso 1: Identificar qu√© se repite y qu√© funcionalidades tenemos

| Funcionalidad | ¬øSe repite? | Funci√≥n a crear |
|---------------|-------------|------------------|
| Mostrar men√∫ | S√≠ (cada iteraci√≥n) | `mostrar_menu()` |
| Pedir dos n√∫meros | S√≠ (4 veces) | `pedir_numeros()` |
| Mostrar resultado | S√≠ (4 veces) | `mostrar_resultado()` |
| Sumar | No | `sumar()` |
| Restar | No | `restar()` |
| Multiplicar | No | `multiplicar()` |
| Dividir | No | `dividir()` |

### Paso 2: Crear las funciones

```python
# ============================================
# VERSI√ìN 2: MODULARIZADA CON FUNCIONES
# ============================================

# ------------------------------
# DEFINICI√ìN DE FUNCIONES
# ------------------------------

def mostrar_menu() -> None:
    """Muestra el men√∫ de opciones de la calculadora"""
    print("")
    print("===== CALCULADORA =====")
    print("1. Sumar")
    print("2. Restar")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Salir")
    print("=======================")


def pedir_numeros() -> tuple:
    """Pide dos n√∫meros al usuario
    
    Returns:
        tuple: (num1, num2) los dos n√∫meros introducidos
    """
    num1 = float(input("N√∫mero 1: "))
    num2 = float(input("N√∫mero 2: "))
    return num1, num2


def mostrar_resultado(num1: float, operador: str, num2: float, resultado: float) -> None:
    """Muestra el resultado de una operaci√≥n con formato
    
    Args:
        num1: Primer operando
        operador: S√≠mbolo de la operaci√≥n (+, -, *, /)
        num2: Segundo operando
        resultado: Resultado de la operaci√≥n
    """
    print("------------------------")
    print(f"{num1} {operador} {num2} = {resultado}")
    print("------------------------")


def sumar(a: float, b: float) -> float:
    """Suma dos n√∫meros"""
    return a + b


def restar(a: float, b: float) -> float:
    """Resta dos n√∫meros"""
    return a - b


def multiplicar(a: float, b: float) -> float:
    """Multiplica dos n√∫meros"""
    return a * b


def dividir(a: float, b: float) -> float:
    """Divide dos n√∫meros
    
    Returns:
        El resultado de la divisi√≥n, o None si b es 0
    """
    if b != 0:
        return a / b
    else:
        return None


# ------------------------------
# PROGRAMA PRINCIPAL
# ------------------------------

opcion = 0

while opcion != 5:
    
    mostrar_menu()
    opcion = int(input("Elige opci√≥n: "))
    
    if opcion >= 1 and opcion <= 4:
        # Todas las operaciones necesitan dos n√∫meros
        num1, num2 = pedir_numeros()
        
        if opcion == 1:
            resultado = sumar(num1, num2)
            mostrar_resultado(num1, "+", num2, resultado)
            
        elif opcion == 2:
            resultado = restar(num1, num2)
            mostrar_resultado(num1, "-", num2, resultado)
            
        elif opcion == 3:
            resultado = multiplicar(num1, num2)
            mostrar_resultado(num1, "*", num2, resultado)
            
        elif opcion == 4:
            resultado = dividir(num1, num2)
            if resultado is not None:
                mostrar_resultado(num1, "/", num2, resultado)
            else:
                print("Error: No se puede dividir entre 0")
                
    elif opcion == 5:
        print("¬°Hasta luego!")
        
    else:
        print("Opci√≥n no v√°lida")
```

### ‚úÖ Ventajas de esta versi√≥n:

1. **No hay c√≥digo repetido**: Cada funcionalidad est√° en UN solo sitio
2. **F√°cil de leer**: El programa principal es muy claro
3. **F√°cil de modificar**: Si quiero cambiar el formato, solo toco UNA funci√≥n
4. **F√°cil de probar**: Puedo probar cada funci√≥n por separado
5. **Reutilizable**: Las funciones se pueden usar en otros programas

---

## üî¨ Probando las funciones por separado

Una ventaja de tener funciones es que podemos probarlas individualmente:

```python
# Probar la funci√≥n sumar
print(sumar(5, 3))       # 8.0
print(sumar(10, -4))     # 6.0
print(sumar(0, 0))       # 0.0
```

```python
# Probar la funci√≥n dividir
print(dividir(10, 2))    # 5.0
print(dividir(10, 0))    # None (no da error!)
print(dividir(7, 3))     # 2.333...
```

```python
# Probar mostrar_resultado
mostrar_resultado(10, "+", 5, 15)
```

## üìù Resumen: Pasos para modularizar

1. **Identificar c√≥digo repetido** ‚Üí Crear una funci√≥n
2. **Identificar bloques con una responsabilidad clara** ‚Üí Crear una funci√≥n
3. **Nombrar las funciones con verbos** (mostrar, pedir, calcular, validar...)
4. **Separar**: Primero funciones, luego programa principal
5. **Documentar**: A√±adir docstrings a cada funci√≥n
